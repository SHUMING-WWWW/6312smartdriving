```python
import matplotlib.pyplot as plt

# Data
models = ['MLP', 'CNN', 'RNN', 'BiRNN', 'LSTM', 'BiLSTM', 'GRU', 'BiGRU']
accuracy = [77.63, 79.92, 78.35, 78.25, 80.66, 80.00, 81.49, 81.52]

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(models, accuracy, color='skyblue')

# Add title and labels
plt.title("Model Test Accuracy Comparison", fontsize=14)
plt.xlabel("Model", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.ylim(75, 83)  # Set y-axis range for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate accuracy values above each bar
for bar, acc in zip(bars, accuracy):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.2,
             f'{acc:.2f}%', ha='center', va='bottom', fontsize=10)

plt.tight_layout()

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data
models = ['BERT', 'RoBERTa', 'DistilRoBERTa']
accuracy = [82.24, 83.80, 83.21]
loss = [0.6966, 0.6676, 0.5609]
training_time = ['2:13:21', '3:22:42', '1:39:12']  # Convert to numerical for plotting

# Convert training time to minutes for plotting
training_time_minutes = [2 * 60 + 13 + 21 / 60, 3 * 60 + 22 + 42 / 60, 1 * 60 + 39 + 12 / 60]
```

```python
# Plot 1: Test Accuracy
plt.figure(figsize=(10, 6))
bars = plt.bar(models, accuracy, color='skyblue')
plt.title("Test Accuracy of Transformer Models", fontsize=14)
plt.xlabel("Model", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.ylim(80, 84)
for bar, acc in zip(bars, accuracy):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{acc:.2f}%', ha='center', va='bottom', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Test Loss
plt.figure(figsize=(10, 6))
bars = plt.bar(models, loss, color='lightcoral')
plt.title("Test Loss of Transformer Models", fontsize=14)
plt.xlabel("Model", fontsize=12)
plt.ylabel("Test Loss", fontsize=12)
plt.ylim(0.5, 0.75)
for bar, ls in zip(bars, loss):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{ls:.4f}', ha='center', va='bottom', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Training Time
plt.figure(figsize=(10, 6))
bars = plt.bar(models, training_time_minutes, color='palegreen')
plt.title("Training Time of Transformer Models", fontsize=14)
plt.xlabel("Model", fontsize=12)
plt.ylabel("Training Time (Minutes)", fontsize=12)
for bar, time, label in zip(bars, training_time_minutes, training_time):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             label, ha='center', va='bottom', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
